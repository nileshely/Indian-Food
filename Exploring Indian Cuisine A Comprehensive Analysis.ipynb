# INDIAN FOOD RECIPES DATASET REPORT

# Introduction:
The Indian Food Recipes dataset is a comprehensive collection of culinary information sourced primarily from Archana's Kitchen website. Comprising 6871 entries, this dataset offers a rich repository of recipes covering a wide array of cuisines, courses, and dietary preferences. Each entry in the dataset includes detailed information such as recipe name, ingredients, preparation and cooking times, servings, cuisine type, course category, dietary specifications, and cooking instructions. Moreover, to ensure accessibility and consistency, the dataset includes both original content and translated versions using the Google Translate API, making it a valuable resource for the analytics community interested in exploring various facets of Indian cuisine.

# Data Overview:
The dataset consists of 15 columns, including a serial number (Srno) for indexing purposes. Key columns such as RecipeName, Ingredients, PrepTimeInMins, CookTimeInMins, TotalTimeInMins, Servings, Cuisine, Course, Diet, Instructions, and their translated counterparts provide essential information for analysis and visualization. Notably, the dataset covers a diverse range of cuisines, including but not limited to North Indian, South Indian, Mughlai, and Continental, catering to varied culinary preferences.

# Exploratory Data Analysis (EDA):
Upon conducting exploratory data analysis, several insights can be gleaned from the dataset. Visualizations reveal the distribution of cuisines, courses, and dietary preferences, enabling a better understanding of the dataset's composition. Additionally, histograms depicting the distribution of preparation, cooking, and total times offer insights into the time requirements for different recipes. Box plots further elucidate the variability in serving sizes across cuisines and the impact of course type on total cooking time.

# Insights and Trends:
Analysis of the dataset highlights several interesting trends and patterns within Indian cuisine. For instance, North Indian cuisine appears to be dominant among the recorded recipes, reflecting its popularity across the country and beyond. Furthermore, vegetarian recipes outnumber non-vegetarian ones, indicating a preference for plant-based diets among consumers. The dataset also sheds light on the diversity of courses, ranging from appetizers and main courses to desserts, catering to various dining occasions and preferences.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/kaggle/input/indian-food-dataset/IndianFoodDatasetCSV.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(15, 6))
df['Cuisine'].value_counts().plot(kind='bar')
plt.title('Distribution of Cuisine')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['PrepTimeInMins'], bins=20, kde=True)
plt.title('Preparation Time Distribution')
plt.xlabel('Time (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['CookTimeInMins'], bins=20, kde=True)
plt.title('Cooking Time Distribution')
plt.xlabel('Time (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalTimeInMins'], bins=20, kde=True)
plt.title('Total Time Distribution')
plt.xlabel('Time (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
numeric_cols = ['PrepTimeInMins', 'CookTimeInMins', 'TotalTimeInMins', 'Servings']
plt.figure(figsize=(10, 8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(x='Cuisine', y='Servings', data=df)
plt.title('Boxplot of Servings by Cuisine')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Course', y='TotalTimeInMins', data=df)
plt.title('Boxplot of Total Time by Course')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.pairplot(df[numeric_cols])
plt.suptitle('Pairplot of Numeric Variables', y=1.02)
plt.show()
```
